```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import approximate_taylor_polynomial
import sympy as sp

# Задана функція
def f(x):
    return np.cos(3*x - 1) + x

# Точка розгляду
x0 = 0.5

# Символьна змінна
x_sym = sp.symbols('x')

# Знаходимо перші три похідні символьно
f_sym = sp.cos(3*x_sym - 1) + x_sym
f_prime_sym = sp.diff(f_sym, x_sym)
f_double_prime_sym = sp.diff(f_prime_sym, x_sym)
f_triple_prime_sym = sp.diff(f_double_prime_sym, x_sym)

# Отримуємо значення похідних у точці розгляду
f_prime_at_x0 = f_prime_sym.subs(x_sym, x0)
f_double_prime_at_x0 = f_double_prime_sym.subs(x_sym, x0)
f_triple_prime_at_x0 = f_triple_prime_sym.subs(x_sym, x0)

# Розрахунок многочлена Тейлора за допомогою approximate_taylor_polynomial
degree = 3
taylor_approximation = approximate_taylor_polynomial(f, x0, degree, scale=1)

# Отримання значень для графіку
x_values = np.linspace(0, 1, 100)
approx_values = taylor_approximation(x_values)

# Графік функції та наближення
plt.figure(figsize=(10, 6))
plt.plot(x_values, f(x_values), label='f(x)')
plt.plot(x_values, approx_values, label='Тейлорське наближення')
plt.scatter(x0, f(x0), color='red', marker='o', label=f'Точка розгляду (x={x0})')
plt.title(f'Тейлорське наближення до f(x) при x={x0}')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# Вивід результатів
print(f"Точка розгляду (x): {x0}")
print(f"f(x): {f(x0)}")
print(f"f'(x): {f_prime_at_x0}")
print(f"f''(x): {f_double_prime_at_x0}")
print(f"f'''(x): {f_triple_prime_at_x0}")
print(f"Тейлорське наближення при x={x0}: {taylor_approximation}")
```
