```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange

# Дані
xi = np.array([-1, 0, 2, 3])
fxi = np.array([-16, -7, -1, 2])

# Точки для обчислення наближених значень
target_points = np.array([-0.5, 0.5, 1.5, 2.5])

# Побудова інтерполяційного багаточлена Лагранжа
poly = lagrange(xi, fxi)

# Обчислення наближених значень у заданих точках
approximated_values = poly(target_points)

# Графік інтерполяційної функції
x_range = np.linspace(min(xi), max(xi), 100)
y_range = poly(x_range)

# Додайте додаткові точки для перевірки
smooth_x_range = np.linspace(min(xi), max(xi), 500)
smooth_y_range = poly(smooth_x_range)

plt.scatter(xi, fxi, color='red', label='Точки')
plt.plot(x_range, y_range, label='Інтерполяційний багаточлен Лагранжа')
plt.scatter(target_points, approximated_values, color='blue', label='Нові точки')
plt.plot(smooth_x_range, smooth_y_range, linestyle='dashed', color='green', label='Перевірка')
plt.title('Графік інтерполяційної функції')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()

# Виведення результатів
print("Наближені значення у заданих точках:")
for x, f_approx in zip(target_points, approximated_values):
    print(f"f({x}) ≈ {f_approx:.3f}")
```
